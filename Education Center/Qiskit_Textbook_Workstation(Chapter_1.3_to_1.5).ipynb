# Representing Qubit States

In this chapter 1.3 we will learn how to represent Qubit states in the **Qiskit** framework for quantum computers/systems. We will understand that these systems are able to process and/or manipulate a larger amount of information given their vector/tensor based operations, but also their ability to operate between spaces of several dimensions.

From the textbook: *" In quantum computers, our basic variable is the qubit: a quantum variant of the bit. These have exactly the same restrictions as normal bits do: they can store only a single binary piece of information, and can only ever give us an output of 0 or 1. However, they can also be manipulated in ways that can only be described by quantum mechanics. This gives us new gates to play with, allowing us to find new ways to design algorithms.
   To fully understand these new gates, we first need to understand how to write down qubit states. For this we will use the mathematics of vectors, matrices, and complex numbers."* 

## Classical vs Quantum bits

In quantum physics basic operations are done with vectors/tensors. The most basic operations are carried out to obtain what is called a *statevector*. In classical mechanics we can also operate with vectors, but usually these only allow for a limited set of operations, a magnitude, a direction or a coordinate pair. But for quantum systems we can extend this to what we call a *state* of a system. In a *state* we represent our vector as a *matrix* of values. This allows for multidimensional operations.

Consider the example from the textbook where we could represent the state of car running in a unidimensional road with some specified direction and instataneous velocity. In a classical represenation of bits this could be a single number (the specific position of the car) giving us the information required: 

  

*" In quantum physics we use statevectors to describe the state of our system. Say we wanted to describe the position of a car along a track, this is a classical system so we could use a number **x**. 


For the quantum way of representing the same information we think of probability of finding the car in a specified position. 

 

*" Alternatively, we could instead use a collection of numbers in a vector called a statevector. Each element in the statevector contains the probability of finding the car in a certain place (...) "*

|x >  = \begin{bmatrix} 0\\ \vdots \\ 0 \\ 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix} \begin{matrix} \ \leftarrow \end{matrix} \begin{matrix} \ \text{Probability of} \\ \text{car being at} \\ \text{position 4} \end{matrix}

*" This isn’t limited to position, we could also keep a statevector of all the possible speeds the car could have, and all the possible colours the car could be. With classical systems (like the car example above), this is a silly thing to do as it requires keeping huge vectors when we only really need one number. But as we will see in this chapter, statevectors happen to be a very good way of keeping track of quantum systems, including quantum computers. "*



### Qubit Notation

In classical computation the basic units of compute are bits. Bits can only be represented as a 0 or 1. This is called a binary rtepresentation of computation. In quantum computing we are not constrained by this limitation. The actual computation in a quantum computing is done in a high dimensional space of vectors/tensors, and when we want or need to know the result of a particular compute process, or a circuit design, we need to make a measurement to extract the output of that process. Only then we are presented with also a binary representation of 0s or 1s, which represent what in quantum physics is called a *collapse of the wavefunction*. For the moment we don't need to be concerned about these *wavefunction collapses* in order to proceed.

We need to understand that the measurement procedure only give us a probability outcome. The actual qubit state is not in all *facts* that particular measurement outcome. The actual quantum system continues to a more complex set of states other than waht we measure. But what we measure is considered *for all practical purposes* the state of the quantum system. As stated in the Qiskit textbook: *" As we saw in the last section, it is possible to prepare a qubit in a state for which it definitely gives the outcome 0 when measured."*  

*" We need a name for this state. Let's be unimaginative and call it  0. Similarly, there exists a qubit state that is certain to output a 1. We'll call this 1. These two states are completely mutually exclusive. Either the qubit definitely outputs a 0, or it definitely outputs a 1. There is no overlap. One way to represent this with mathematics is to use two orthogonal vectors."* :



\begin{aligned}
|0\ > = \begin{bmatrix} 1 \\ 0 \end{bmatrix}   \\   |1\ > = \begin{bmatrix} 0 \\ 1 \end{bmatrix}.
\end{aligned}

The representation above is what we call the qubit notation and it is the standard scientific practice notation. It derives from the **Dirac notation**, named by physicist **Paul Dirac**, and the representation above describes a *ket* 0 and a *ket* 1. The other part of the notation could describe what is called a *bra* 0 or a *bra* 1. For this reason it is also called the *BraKet* notation. The *bra* is represented as the *ket* inverted, but it actually is a row vector representation (<0| or <1|). The *ket* is a column vector representation. So, those numbers are vectors (or tensors of rank 1).

A good quote from the Qiskit textbook about these vector representations. (Note that they are not the same as the vector representations we learn at high school basic calculus lessons): 

*" If you are not familiar with vectors, you can essentially just think of them as lists of numbers which we manipulate using certain rules. If you are familiar with vectors from your high school physics classes, you'll know that these rules make vectors well-suited for describing quantities with a magnitude and a direction. For example, the velocity of an object is described perfectly with a vector. However, the way we use vectors for quantum states is slightly different to this, so don't hold on too hard to your previous intuition. It's time to do something new! "*

With this notation we can represent more complex qubit states than the basic |0> or |1>. For example:

\begin{aligned}
|q_0\ > = \begin{bmatrix} \tfrac{1}{\sqrt{2}} \\ \tfrac{i}{\sqrt{2}} \end{bmatrix} .
\end{aligned}

*" To understand what this state means, we'll need to use the mathematical rules for manipulating vectors. Specifically, we'll need to understand how to add vectors together and how to multiply them by scalars. "*

This |q_0> state can be represented as a linear combination of the former |0> and |1>. This is so because |0> and |1> form an important algebra set of vectors: an orthonormal basis. With an orthonormal basis we can represent any other vector from a high-dimensional space, with any complexity as we like or need:

\begin{aligned}
|q_0\ > = \begin{matrix}\tfrac{1}{\sqrt{2}}|0\ >  \end{matrix}  + \begin{matrix} \tfrac{i}{\sqrt{2}}|1\ > \end{matrix}.
\end{aligned}

*" This vector,  |q0⟩ is called the qubit's statevector, it tells us everything we could possibly know about this qubit. For now, we are only able to draw a few simple conclusions about this particular example of a statevector: it is not entirely |0⟩ and not entirely |1⟩. Instead, it is described by a linear combination of the two. In quantum mechanics, we typically describe linear combinations such as this using the word 'superposition'. "*

*" Though our example state  |q0⟩  can be expressed as a superposition of  |0⟩  and  |1⟩ , it is no less a definite and well-defined qubit state than they are. To see this, we can begin to explore how a qubit can be manipulated. "*

# The Rules of Measurement

In Quantum Computing there are rules to follow. Some of them are very important ones. That is the case with rules to perform a measurement of a qubit(s) system(s). In quantum physics we've learned that a mewasurement of a particular state involves computing a probability outcome. That is also the case in quantum computing. Here we will compute the probability of measuring the outcome of a qubit operation over a wavefunction \begin{aligned}|\psi\ > \end{aligned} which we calculate(operate) to be in state \begin{aligned} |x\ > \end{aligned}. We do:

\begin{aligned} p(|x >)  = | < |x | \psi >|^2
\end{aligned}


As stated in the textbook: *" In the equation above, |x > can be any qubit state. To find the probability of measuring |x >, we take the inner product of |x > and the state we are measuring (in this case |ψ >), then square the magnitude.

A reminder of the rules of the inne product: < a|  = \begin{bmatrix}a_0^*, & a_1^*, & \dots & a_n^* \end{bmatrix}, 
    |b >  = \begin{bmatrix}b_0 \\ b_1 \\ \vdots \\ b_n \end{bmatrix}

< a|b > = \begin{matrix} a_0^* b_0  +  a_1^* b_1  ...  a_n^* b_n \end{matrix}

*" We can see that the inner product of two vectors always gives us a scalar. A useful thing to remember is that the inner product of two orthogonal vectors is 0, for example if we have the orthogonal vectors |0 > and |1 > "* :


\begin{aligned}
      < 1|0 > = \begin{bmatrix} 0 , & 1\end{bmatrix}\begin{bmatrix}1 \\ 0\end{bmatrix} = 0
       \end{aligned}

*" Additionally, remember that the vectors  |0 > and |1 > are also normalised (magnitudes are equal to 1) "*:

\begin{aligned} 
      < 0|0 > & = \begin{bmatrix} 1 , & 0\end{bmatrix}\begin{bmatrix}1 \\ 0\end{bmatrix} = 1 \\
      < 1|1 > & = \begin{bmatrix} 0 , & 1\end{bmatrix}\begin{bmatrix}0 \\ 1\end{bmatrix} = 1
      \end{aligned}

We can now look at the state  \begin{matrix} |q_0\ > \end{matrix} from before and verify that indeed that the probability of measuring |0 > is \begin{matrix} 0.5 \end{matrix} :  

\begin{aligned}
|q_0 > & = \tfrac{1}{\sqrt{2}}|0 > + \tfrac{i}{\sqrt{2}}|1 > \\
< 0| q_0 >  & = \tfrac{1}{\sqrt{2}}< 0|0 > + \tfrac{i}{\sqrt{2}}< 0|1 > \\
& = \tfrac{1}{\sqrt{2}}\cdot 1 +  \tfrac{i}{\sqrt{2}} \cdot 0\\
& = \tfrac{1}{\sqrt{2}}\\
|< 0| q_0 >|^2 & = \tfrac{1}{2}
\end{aligned}

This rule governs how we get information out of quantum states. It is therefore very important for everything we do in quantum computation. This as several implications and we discuss below some of them.

### Normalisation 

*" The rule shows us that amplitudes are related to probabilities. If we want the probabilities to add up to 1 (which they should!), we need to ensure that the statevector is properly normalized. Specifically, we need the magnitude of the state vector to be 1. "* :

\begin{aligned}
< \psi|\psi >  = 1 \\
\end{aligned}

Implying:

\begin{aligned}
|\psi > = \alpha|0 > + \beta|1 >  
\end{aligned}

Then:

\begin{aligned} 
\sqrt{|\alpha|^2 + |\beta|^2} = 1
\end{aligned}

*" This explains the factors of \begin{matrix} \sqrt{2} \end{matrix} you have seen throughout this chapter. In fact, if we try to give *initialize()* a vector that isn’t normalised, it will give us an error "* :

```python
vector = [1,1]
qc.initialize(vector, 0)
```

### Alternative Measurement

AS the quantum computing space its operators function is high-dimensional, the measurements to know the particular state, that is ist probabability \begin{aligned} p(|x \ >) \end{aligned} can also take an infinte number of possibiliteis. There no mandatory rule prescribing that the outcome of measuremtns should be \begin{aligned} |0 \ > \ or \ |1 \ > \end{aligned}.

*" The measurements we have considered so far are in fact only one of an infinite number of possible ways to measure a qubit. For any orthogonal pair of states, we can define a measurement that would cause a qubit to choose between the two. (...) For now, just bear in mind that \begin{aligned} **|x  >** is not limited to being simply **|0  >** or  **|1  >**. "* 

### Global Phase

When we measure the state **|1 >** we know for sure that the we will have the output **1** with certainty. But we can also represent other states, such as:

\begin{aligned}
\begin{bmatrix} 0 \\ i \end{bmatrix} = i|1 >
\end{aligned}

To see how this behaves, we apply the measurement rule.

**\begin{aligned}
|<x| (i|1 >) |^2 = | i > x|1 >|^2 = |<x|1 >|^2
\end{aligned}** 

Here we can easily understand that the complex number *i* disapears as soon as we take the magnitude (absolute value) of the inner product squared.

*" It does not matter what measurement we are considering, the probabilities for the state **i|1 >** are identical to those for **|1 >** . Since measurements are the only way we can extract any information from a qubit, this implies that these two states are equivalent in all ways that are physically relevant."*

** *More generally, we refer to any overall factor γ on a state for which |γ| = 1 as a 'global phase'. States that differ only by a global phase are physically indistinguishable.* **

\begin{aligned}
|< x| ( \gamma |a >) |^2 = | \gamma < x|a >|^2 = |< x|a >|^2
\end{aligned}

It is important to note here that this phase is different from the ** *relative phase***, which is a phase *difference* between terms in superposition. Once we study multiple qubits operations and multiple types of measurements this issue will become relevant again.

### The Observer Effect

When we measure the amplitudes of a signal from a qubit (the only way we have really to actual *measure* a qubit state) we know that we are extrating information about the probability of us finding that qubit in a particular state. But **after** the measurement we know for sure what the state of the qubit is. For example:

\begin{aligned}
|q > = \alpha|0 > + \beta|1 >
\end{aligned}

This qubit is in the state **|0 >**. And if we measure it again there is a 100% chance of finding the qubit in the state **|0 >**. What this means is the from a more general abstract qubit state, once we do a measurement we change the original state of an up until then unknown quibt state:

\begin{aligned}
|q > = \begin{bmatrix} \alpha \\ \beta \end{bmatrix} \xrightarrow{\text{Measure }|0 >} |q > = |0 > = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
\end{aligned}

This is an important passage from the textbook worth to highlight: "* **We sometimes refer to this as collapsing the state of the qubit. It is a potent effect, and so one that must be used wisely. For example, were we to constantly measure each of our qubits to keep track of their value at each point in a computation, they would always simply be in a well-defined state of either**  **|0 >** or **|1 >**. ** As such, they would be no different from classical bits and our computation could be easily replaced by a classical computation. To achieve truly quantum computation we must allow the qubits to explore more complex states. Measurements are therefore only used when we need to extract an output. This means that we often place all the measurements at the end of our quantum circuit.** *" 


*" We can demonstrate this using **Qiskit**’s statevector simulator. Let's initialise a qubit in superposition: "



```python
qc = QuantumCircuit(1) # We are redefining qc
initial_state = [0.+1.j/sqrt(2),1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
qc.draw()
```

This initializes the qubit in the state:

\begin{aligned}
|q > = \tfrac{i}{\sqrt{2}}|0 > + \tfrac{1}{\sqrt{2}}|1 >
\end{aligned}

Using the *statevector* simulator we can verify this:

```python
state = execute(qc, backend).result().get_statevector()
print("Qubit State = " + str(state))
```

*" We can see here the qubit is initialised in the state [0.+0.70710678j 0.70710678+0.j], which is the state we expected.
Let’s now measure this qubit: *"

```python
qc.measure_all()
qc.draw()
```

*" When we simulate this entire circuit, we can see that one of the amplitudes is always 0: "*

```python
state = execute(qc, backend).result().get_statevector()
print("State of Measured Qubit = " + str(state))
```

# The Block Sphere

## The Restricted Qubit State

As we've seen the general abstract state of a qubit (**|q >**) is represented by both \begin{aligned} {\alpha}, \ {\beta} \end{aligned} complex valued numbers: 

\begin{aligned}
|q > = \alpha|0 > + \beta|1 > \\
\alpha, \beta \in \mathbb{C}
\end{aligned}

"* The first two implications in section 2 tell us that we cannot differentiate between some of these states. This means we can be more specific in our description of the qubit. Firstly, since we cannot measure global phase, we can only measure the difference in phase between the states **|0 >** and **|1 >** . Instead of having α and  β be complex, we can confine them to the real numbers and add a term to tell us the relative phase between them: *

 \begin{aligned}
 |q > = \alpha|0 > + e^{i\phi}\beta|1 > \\
 \alpha, \beta, \phi \in \mathbb{R}
 \end{aligned}
 

The normalization yields us a magnitude of 1 for the qubit **|q >** from the parameters \begin{aligned} \alpha \ and \  \beta \end{aligned}:

\begin{aligned}
\sqrt{\alpha^2 + \beta^2} = 1
\end{aligned}

Using a well known trigonometric identity and describing the above parameters in terms of one variable, \begin{aligned} \theta \end{aligned}:

\begin{aligned}
\alpha = \cos{\tfrac{\theta}{2}}, \quad \beta=\sin{\tfrac{\theta}{2}}
\end{aligned}

The two angles/variables \begin{aligned} \phi \ and \  \theta \end{aligned} can now be used to describe the state of any qubit:

\begin{aligned}
|q > = \cos{\tfrac{\theta}{2}}|0 > + e^{i\phi}\sin{\tfrac{\theta}{2}}|1 > \\
\theta, \phi \in \mathbb{R}
\end{aligned}

**Qiskit** allow us to plot this general qubit state:

\begin{aligned}
|q > = \cos{\tfrac{\theta}{2}}|0 > + e^{i\phi}\sin{\tfrac{\theta}{2}}|1 >
\end{aligned}


" **If we interpret θ and ϕ as spherical co-ordinates (r = 1, since the magnitude of the qubit state is 1), we can plot any single qubit state on the surface of a sphere, known as the *Bloch Sphere*.** "

" * Below we have plotted a qubit in the state |+ >. In this case, θ = π/2 and ϕ = 0* "

```python
from qiskit_textbook.widgets import plot_bloch_vector_spherical
coords = [pi/2,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

## The Pauli Gates

We continue now into the section 1.4 of the **Qiskit Textbook**. In this section we will study and describe the single-qubit unit of computation and their fundamental operations: gates. This prepares the setting for the next chapter where the Multi-Qubit case for quantum computation will be studied. We start with undesrtanding and explaining the Pauli Gate.

But first let us get a glimpse of the actual physical meaning of quantum circuits and their operations we are going to describe, their gates. From the Textbook we could read: 

"* In The Atoms of Computation we came across some gates and used them to perform a classical computation. An important feature of quantum circuits is that, between initialising the qubits and measuring them, the operations (gates) are always reversible! These reversible gates can be represented as matrices, and as rotations around the Bloch sphere. *"

This can be visualized using this code:

```python
from qiskit import *
from math import pi
from qiskit.visualization import plot_bloch_multivector
```

After reminding ourselves the Linear Albebra origins of the Pauli matrices (they are basically 2x2 complex valued symmetric matrices - *Hermitian Matrix*) we will explain now how they perform when computing. In here we begin to understand the operations of gates on quantum circuits. First the X-Gate:

### The X-Gate

"* The X-gate is represented by the Pauli-X matrix:*"

\begin{aligned}
X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} = |0\ >\ < 1| + |1\ >\ < 0|
\end{aligned}

This operation is an Amplitude Switcher: "* To see the effect a gate has on a qubit, we simply multiply the qubit’s statevector by the gate. We can see that the X-gate switches the amplitudes of the states |0 > and |1 >: *"

\begin{aligned}
X|0\ > = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}\begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \end{bmatrix} = |1\ >
\end{aligned}

Qiskit provides a *short-circuit* for a verification of this operation:


```python
# Let's do an X-gate on a |0> qubit
qc = QuantumCircuit(1)
qc.x(0)
qc.draw()
```

This circuit can also be checked for its result. For this we go back to the Aer backend and use the *statevector_simulator* and the output as argument inot *plot_block_multivector*:

```python
# Let's see the result
backend = Aer.get_backend('statevector_simulator')
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

The operation of the X-Gate is thus analogous to the classical *NOT-Gate*, but here we are dealing with operations of high-dimensional complex vector spaces and with angles of rotation. Not a simple bit-flip.

"* We can indeed see the state of the qubit is |1> as expected. We can think of this as a rotation by π radians around the x-axis of the Bloch sphere. The X-gate is also often called a NOT-gate, referring to its classical analogue.*"

### The Y & Z-gates

Similar to the X-Gate operatoin on the X axis of the Block Sphere, we can do the same for the Y & Z axis, retrieving the Y & Z-Gates. The respective matrices entries are diferent, with Y-Gate having complex valued entries:

\begin{aligned}
Y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} \quad\quad\quad\quad Z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
\end{aligned}

\begin{aligned}
Y = -i|0\ >\ < 1| + i|1\ >\ < 0| \quad\quad Z = |0\ >\ < 0| - |1\ >\ < 1|
\end{aligned}

"* And, unsurprisingly, they also respectively perform rotations by π around the y and z-axis of the Bloch sphere. *" 

"* In Qiskit, we can apply the Y and Z-gates to our circuit using: *"



```python
qc.y(0) # Do Y-gate on qubit 0
qc.z(0) # Do Z-gate on qubit 0
qc.draw()
```

### Digression: The X, Y & Z-Bases 


The Z-Gate has a special propety in the context of Qiskit and Quantum Computing conventions generally. It forms what's called a *computational basis* (the most basic computing basis), and indeed it does not change anything of the original input. This is because the states |0> and |1> are the two *eigenstates* of the Z-gate. For this fact sometimes the computational basis in quantum computing is also called a Z-basis. "* This is not the only basis we can use, a popular basis is the X-basis, formed by the eigenstates of the X-gate. We call these two vectors | + > and | - >: *"

\begin{aligned}
|+\ > = \tfrac{1}{\sqrt{2}}(|0\ > + |1\ >) = \tfrac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}
\end{aligned}
\begin{aligned}
|-\ > = \tfrac{1}{\sqrt{2}}(|0\ > - |1\ >) = \tfrac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}
\end{aligned}

From the Textbook: "* Another less commonly used basis is that formed by the eigenstates of the Y-gate. These are called: *"

\begin{aligned}
|\circlearrowleft\ >, \quad |\circlearrowright\ >
\end{aligned}

"* There are in fact an infinite number of bases; to form one, we simply need two orthogonal vectors. *"



## The Hadamard Gate 


In order to go now to meore interesting and computationally significant operations we need to put *Qubits* in **superposition**. For that we need another type of Gate, beyond the Pauli Gates (Pauli Gates can only do operations on |0> and |1>). The first such Gate is the **Hadamard Gate** : 

\begin{aligned}
H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1   &  1  \\  1  & -1 \end{bmatrix}
\end{aligned}

*source: https://en.wikipedia.org/wiki/Quantum_logic_gate#Hadamard_(H)_gate.*

What are the transformations the Hadamard Gate performs. These are shown below:

\begin{aligned}
H|0\ > = |+\ >
\end{aligned}
\begin{aligned}
H|1\ > = |-\ >
\end{aligned}

Textbook: "* This can be thought of as a rotation around the Bloch vector [1,0,1] (the line between the x & z-axis), or as transforming the state of the qubit between the X and Z bases. *" 

### Digression: Measuring in Different Bases

Now an important digression into the topic of **measurement** in Quantum Computing. This is an all-important operation for quantum computing. The measurement operation allow us to obtaiun the result of our desired computation. And for those with a more in-deep knowledge of Quantum Physics, also know that measuring a quantum system changes the original input, which we assumed to ahve been in superposition or *entangled* in some way. We call this the *collapse of the wavefunction*, but for Quantum Computing it also means the **output of computation**.

From the Qiskit Textbook we can read: "* We have seen that the Z-axis is not intrinsically special, and that there are infinitely many other bases. Similarly with measurement, we don’t always have to measure in the computational basis (the Z-basis), we can measure our qubits in any basis. As an example, let’s try measuring in the X-basis. We can calculate the probability of measuring either |+ > or |- >: *"

\begin{aligned}
p(|+\ >) = |\ < +|q\ >|^2, \quad p(|-\ >) = |\ < -|q\ >|^2
\end{aligned}

"* And after measurement, we are guaranteed to have a qubit in one of these two states. *"
But Qiskit only allows measurung in the Z-basis. Therefore we need to create our own measurement basis using Hadamard Gates. In order to achieve this we need to define a function appropriate for this task. Below we now show the code for this operation (with initialization and the measurement visualization):

```python
# Create the X-measurement function:
def x_measurement(qc,qubit,cbit):
    """Measure 'qubit' in the X-basis, and store the result in 'cbit'"""
    qc.h(qubit)
    qc.measure(qubit, cbit)
    qc.h(qubit)
    return qc

initial_state = [0,1]
# Initialise our qubit and measure it
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
x_measurement(qc, 0, 0)  # measure qubit 0 to classical bit 0
qc.draw()
```

With Qiskit Textbook proper readership we can understabd that we can create an X-gate by *sandwiching* a Z-gate between two Hadamard Gates H-gates:

                                     X = HZH

"* Starting in the Z-basis, the H-gate switches our qubit to the X-basis, the Z-gate peforms a NOT in the X-basis, and the final H-gate returns our qubit to the Z-basis.*" Here it is a good illustration:

"* We can verify this always behaves like an X-gate by multiplying the matrices: *"



\begin{aligned}
HZH =
\tfrac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
\tfrac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
=
\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}
=X
\end{aligned}

"* Following the same logic, we have created an X-measurement by sandwiching our Z-measurement between two H-gates. *"



```python
backend = Aer.get_backend('statevector_simulator') # Tell Qiskit how to simulate our circuit
out_state = execute(qc,backend).result().get_statevector() # Do the simulation, returning the state vector
plot_bloch_multivector(out_state) # Display the output state vector
```

Important remark to read in the Textbook: "* **We initialised our qubit in the state |1 >, but we can see that, after the measurement, we have collapsed our qubit to the states | + > or | − >. If you run the cell again, you will see different results, but the final state of the qubit will always be |+ > or |− >.** *"

Plus this one: "* **Measuring in different bases allows us to see Heisenberg’s famous uncertainty principle in action. Having certainty of measuring a state in the Z-basis removes all certainty of measuring a specific state in the X-basis, and vice versa. A common misconception is that the uncertainty is due to the limits in our equipment, but here we can see the uncertainty is actually part of the nature of the qubit.** *"

"* For example, if we put our qubit in the state |0 >, our measurement in the Z-basis is certain to be |0 >, but our measurement in the X-basis is completely random! Similarly, if we put our qubit in the state | - >, our measurement in the X-basis is certain to be | - >, but now any measurement in the Z-basis will be completely random.
More generally: **Whatever state our quantum system is in, there is always a measurement that has a deterministic outcome.** *"

## The Rϕ-gate

In **Qiskit** and other Quantum Computing projects, we can perform different types of gates has we have seen above. That is the case with the gate we are describing in this section the *Rϕ-gate*. It is a *parametrized* gate, meaning it needs an input number (usually a real number) to specify what it will do. " *The Rϕ-gate performs a rotation of ϕ around the Z-axis direction (and as such is sometimes also known as the Rz-gate). It has the matrix:* "

\begin{aligned}
R_\phi = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\phi} \end{bmatrix}
\end{aligned}

## The I, S and T-gates

### The I-gate

The *I-gate* is also called the *Identity Gate*. As it is it performs the neutral multiplication operation of identity, or multiply by 1, but here the 1 is a symmetric 2x2 matrix. That is, it does not change anything of the original input:

\begin{aligned}
I = \begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix}
\end{aligned}

From the Textbook we can read why this is even a considered important gate to learn and apply. It is actually often applied in the two cases described: *Applying the identity gate anywhere in your circuit should have no effect on the qubit state, so it’s interesting this is even considered a gate. There are two main reasons behind this, one is that it is often used in calculations, for example: proving the X-gate is its own inverse:*

\begin{aligned}
I = XX
\end{aligned}

*The second, is that it is often useful when considering real hardware to specify a ‘do-nothing’ or ‘none’ operation.*

### The S-gates

*The next gate to mention is the S-gate (sometimes known as the √Z-gate), this is an Rϕ-gate with ϕ = π/2. It does a quarter-turn around the Bloch sphere. It is important to note that unlike every gate introduced in this chapter so far, the S-gate is not its own inverse! As a result, you will often see the S†-gate, (also “S-dagger”, “Sdg” or √Z†-gate). The S†-gate is clearly an Rϕ-gate with ϕ = −π/2:*

\begin{aligned}
S = \begin{bmatrix} 1 & 0 \\ 0 & e^{\frac{i\pi}{2}} \end{bmatrix}, \quad  S^\dagger = \begin{bmatrix} 1 & 0 \\ 0 & e^{-\frac{i\pi}{2}} \end{bmatrix}
\end{aligned}

*The name " √Z-gate" is due to the fact that two successively applied S-gates has the same effect as one Z-gate:*

\begin{aligned}
SS|q\ > = Z|q\ >
~\end{aligned}

***To add an S-gate in Qiskit:***



```python
qc = QuantumCircuit(1)
qc.s(0)   # Apply S-gate to qubit 0
qc.sdg(0) # Apply Sdg-gate to qubit 0
qc.draw()
```

### The T-gate

*The T-gate is a very commonly used gate, it is an Rϕ-gate with ϕ = π/4:*
 

\begin{aligned}
T = \begin{bmatrix} 1 & 0 \\ 0 & e^{\frac{i\pi}{4}} \end{bmatrix}, \quad  T^\dagger = \begin{bmatrix} 1 & 0 \\ 0 & e^{-\frac{i\pi}{4}} \end{bmatrix}
\end{aligned}

*As with the S-gate, the T-gate is sometimes also known as the 4^√Z-gate.*


***In Qiskit***:

```python
qc = QuantumCircuit(1)
qc.t(0)   # Apply T-gate to qubit 0
qc.tdg(0) # Apply Tdg-gate to qubit 0
qc.draw()
```

## General U-gates 

What was shown in the cells above were instance gates of the more general *Rϕ-gate*. In the same mane we can now introduce a generalized version of a U-gate (stands for *Unitary gate*), the *U3-gate*. This gate is the most general of all single-qubit quantum gates. It is a parametrised gate of the form:

\begin{aligned}
U_3(\theta, \phi, \lambda) = \begin{bmatrix} \cos(\theta/2) & -e^{i\lambda}\sin(\theta/2) \\
            e^{i\phi}\sin(\theta/2) & e^{i\lambda+i\phi}\cos(\theta/2)
     \end{bmatrix}
\end{aligned}

As we can read from the Textboo, we never see this type of representation of a gate operation, despite the fact that any of the gates we've written so far could be in tis way represented. It certainly has to do with the cumbersome looking mathematical notation of this representation. but this should aid in understanding, as we are always presented with simplified versions of the true represnetations of what is going on, and we have a false sensation of *understanding*...

Reading further form the Textbook: *Qiskit provides U2 and U1-gates, which are specific cases of the U3 gate in which \begin{aligned} \theta = \tfrac{\pi}{2} \end{aligned}, and \begin{aligned} \theta = \phi = 0 \end{aligned} repectively. *

*You will notice that the U1-gate is equivalent to the Rϕ-gate*:


\begin{aligned}
U_3(\tfrac{\pi}{2}, \phi, \lambda) = U_2 = \tfrac{1}{\sqrt{2}}\begin{bmatrix} 1 & -e^{i\lambda} \\
            e^{i\phi} & e^{i\lambda+i\phi}
     \end{bmatrix}
& \quad &
U_3(0, 0, \lambda) = U_1 = \begin{bmatrix} 1 & 0 \\
            0 & e^{i\lambda}\\
     \end{bmatrix}
\end{aligned}

***Before running on real IBM quantum hardware, all single-qubit operations are compiled down to U1, U2 and U3. For this reason they are sometimes called the physical gates.***
